```python
## 8.1
def display_message():
    print("I am learning Python")

display_message()

## 8.2
def favorite_book(titles):
    print(f'my favorite books is {titles.title()}')

favorite_book('Alice in Wonderland')


## 8.3

def make_shirt(size,txt):
    print(f'The size of the shirt is {size} and the text on shirt is {txt.title()}')

make_shirt(40,'Gumel')

make_shirt(txt = 'Jigawa',size='Medium')

## 8.4
def mod_make_shirt(size,txt='I love Python'):
    print(f'The size of the shirt is {size} and the text on shirt is {txt.title()}')
mod_make_shirt('Large')
mod_make_shirt('Medium')

def mod_make_shirt(size,txt='Modified function'):
    print(f'The size of the shirt is {size} and the text on shirt is {txt.title()}')
    print('I love Python')
mod_make_shirt('Small')


## 8.5 
def describe_city(name, country='Nigeria'):
    print(f'{name.title()} is in {country.title()}')
describe_city('Abuja')
describe_city('Kano')
describe_city('Riyadh')




## 8.6
def city_country(city,country):
    cities = f'{city} {country}'
    return cities.title()

first = city_country('Abuja','Nigeria')
print(first)
second = city_country('Accra','Ghana')
print('\n',second)
third = city_country('Dutse','Nigeria')
print('\n',third)

## 8.7
def make_album(artist,title):
    dictionary = {'artist_name':artist,'album_title':title}
    return dictionary
lil_wayne = make_album('Lil_Wayne','Runaway')
print(lil_wayne)
Mayorkun = make_album('Mayorkun','Holy_Father')
print(Mayorkun)
Ayra = make_album('Ayra','Samaritan')
print(Ayra)

def mod_make_album(artist,title,songs=None):
    if songs:
        dictionary = {'artist_name':artist,'album_title':title,'number_of_songs':songs}
    else:
        dictionary = {'artist_name':artist,'album_title':title}
    return dictionary

Ayra = mod_make_album('Ayra','Samaritan',465)
print(Ayra)

## 8.8
def make_album(artist,title):
    dictionary = {'artist_name':artist,'album_title':title}
    return dictionary
while True:
 artist = input("\nPlease Enter name of the artists:")
 if artist == 'q':
    break
 title = input("\nPlease Enter title of the song:")
 if title == 'q':
    break
 entry = make_album(artist,title) 

 print(entry)

```
